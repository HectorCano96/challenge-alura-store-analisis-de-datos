#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
'''
NOTA:
1. DEJO ESTA PRIMER IDEA ARRIBA POR FINES EDUCATIVOS, PERO EL CÓDIGO PRINCIPAL ES EL SEGUNDO.
2. ESTE CÓDIGO ES POCO ESCALABLE Y MÁS REPETITIVO, CONSIDERO QUE SI SE AGREGARAN MÁS TIENDAS DESPUÉS, POR EJEMPLO, SERÍA MÁS COMPLICADO.
3. EL RESULTADO ES EL MISMO: [np.float64(1150880400.0), np.float64(1116343500.0), np.float64(1098019600.0), np.float64(1038375700.0)] -> NO CONVERTIDO A FLOAT, SE VE ASÍ POR EL MÓDULO 'pandas'

# Se define la función para sumar los df y se colocan raises para controlar los errores
def sumar_columna_df(df: pd.DataFrame, nombre_columna: str) -> float:
    """
    Suma los valores numéricos de una columna específica de un DataFrame.

    Args:
        df: DataFrame de pandas que contiene los datos.
        nombre_columna: Nombre de la columna que se desea sumar.

    Returns:
        La suma total de los valores numéricos de la columna especificada.

    Raises:
        KeyError: Si la columna especificada no existe en el DataFrame.
        TypeError: Si la columna contiene valores no numéricos.
    """
    if nombre_columna not in df.columns:
        raise KeyError(f"La columna '{nombre_columna}' no existe en el DataFrame.")

    if not pd.api.types.is_numeric_dtype(df[nombre_columna]):
        raise TypeError(f"La columna '{nombre_columna}' no contiene valores numéricos.")

    return df[nombre_columna].sum()

# Devuelve la suma (un número)
suma_tienda = suma_columna_df(tienda, 'Precio')
suma_tienda2 = suma_columna_df(tienda2, 'Precio')
suma_tienda3 = suma_columna_df(tienda3, 'Precio')
suma_tienda4 = suma_columna_df(tienda4, 'Precio')

sumas_tiendas = [suma_tienda, suma_tienda2, suma_tienda3, suma_tienda4]

print(sumas_tiendas)
'''
```

```python
# Función de formato
def imprimir_elemento(nombre_elemento, cantidad, ancho_total=40):
    texto_elemento = f"{nombre_elemento}: "
    longitud_texto = len(texto_elemento)
    ancho_numero = ancho_total - longitud_texto
    if ancho_numero < 5:  # Evita que sea muy pequeño y se vea mal
        ancho_numero = 5
    print(f"\t{texto_elemento}{cantidad:>{ancho_numero}}")

#Se define una función para almacenar los data frames en un diccionario y así ahorrar código
def crear_diccionario_de_tiendas(**kwargs):
  """
  Crea un diccionario donde las llaves son los nombres dados
  y los valores son los DataFrames correspondientes.

  Args:
    **kwargs: pares nombre=DataFrame

  Returns:
    dict: Diccionario con nombre de tienda como llave y DataFrame como valor.
  """
  return kwargs

#Se asignan las llaves y los df como argumentos
tiendas= crear_diccionario_de_tiendas(
    tienda_1=tienda,
    tienda_2=tienda2,
    tienda_3=tienda3,
    tienda_4=tienda4,
)

# Diccionario para almacenar la información por tienda
info_por_tienda = {}

for nombre_tienda, df in tiendas.items():
  # Validar que exista la columna 'Precio' y sea numérica. Se realiza suma.
  if 'Precio' in df.columns and pd.api.types.is_numeric_dtype(df['Precio']):
    suma_precios = float(df['Precio'].sum())

    # Agregar al diccionario info_por_tienda
    if nombre_tienda in info_por_tienda.items():
      info_por_tienda[nombre_tienda]['ingresos'] = suma_precios
    else:
      info_por_tienda[nombre_tienda] = {'ingresos': suma_precios}
  else:
    if nombre_tienda not in info_por_tienda:
      info_por_tienda[nombre_tienda] = {}

for tienda, datos in info_por_tienda.items():
  print(f"\nIngresos totales en {tienda}:\n")

  ingresos= datos.get('ingresos', {})
  imprimir_elemento('Total', ingresos, ancho_total=40)
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
for nombre_tienda, df in tiendas.items():
  if 'Categoría del Producto' in df.columns:
    # Si se encuentra la columna, iniciar conteo
    conteo = df['Categoría del Producto'].value_counts().to_dict()

    if nombre_tienda in info_por_tienda:
      info_por_tienda[nombre_tienda]['conteo'] = conteo
    else:
      info_por_tienda[nombre_tienda] = {'conteo': conteo}
  else:
    if nombre_tienda not in info_por_tienda:
      info_por_tienda[nombre_tienda] = {}
    # También se puede trabajar como serie en pandas; pd.Series(dtype=int) si no se pasa a diccionario y se queda como serie, para arrojar una serie vacía.
    #---> Se crea una serie de pandas vacía porque None no es compatible con pd.concat() para concatenar las series en caso de ser necesario
'''
#Este bloque visualiza los datos organizados en forma de series, sólo si no se uso .to_dict() y se dejaron como series de pandas
for tienda, conteo in info_por_tienda.items():
  print(f"\nCategorías más populares en {tienda}:")
  if conteo is not None:
    print(conteo.sort_values(ascending=False))
  else:
    print("No se encontró la columna 'Categoría del Producto'")
'''

for tienda, datos in info_por_tienda.items():
    print(f"\nCategorías más populares en {tienda}:\n")

    categorias_conteo = datos.get('conteo', {})

    for categoria, cantidad in sorted(categorias_conteo.items(), key=lambda x: x[1], reverse=True):
          imprimir_elemento(categoria, cantidad, ancho_total=40)
```

```python
# Cálculo del total de ingresos por categoría
for nombre_tienda, df in tiendas.items():
  if 'Precio' in df.columns and 'Categoría del Producto' in df.columns:
    ventas = df.groupby('Categoría del Producto')['Precio'].sum().to_dict()
    # Si ya existe un diccionario con conteo, solo agregamos 'ventas'
    if nombre_tienda in info_por_tienda:
      info_por_tienda[nombre_tienda]['ventas'] = ventas
    else:
      # Si no existe la tienda en el diccionario, la creamos con ventas
      info_por_tienda[nombre_tienda] = {'ventas': ventas}
  else:
    # Si no hay columna 'Precio' o 'Categoría', aseguramos que exista la tienda con ventas vacías
    if nombre_tienda not in info_por_tienda:
      info_por_tienda[nombre_tienda] = {'ventas': {}}

# Visualizando los datos
for tienda, datos in info_por_tienda.items():
  print(f"\nVentas totales por categoría en {tienda}:\n")

  # Accedemos solo a las ventas (puede no existir, por eso uso get con {} por defecto)
  ventas = datos.get('ventas', {})

  for categoria, total in sorted(ventas.items(), key=lambda x: x[1], reverse=True):
    imprimir_elemento(categoria, f"{total: ,.2f}", ancho_total=50)
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
#En este ejercicio agrego un promedio ponderado; el peso se asigna a cada categoría según se impacto en los ingresos generales por tienda para evaluar las calificación según
#su impacto en el ingreso.

for nombre_tienda, df in tiendas.items():
  if all(col in df.columns for col in ['Calificación', 'Categoría del Producto', 'Precio']):

    # Promedio general
    promedio_general = float(df['Calificación'].mean())

    # Promedio por categoría
    promedio_por_categoria = df.groupby('Categoría del Producto')['Calificación'].mean()

    # Suma total de ventas para pesos
    ventas_por_categoria = df.groupby('Categoría del Producto')['Precio'].sum()
    ventas_totales = ventas_por_categoria.sum()
    pesos = ventas_por_categoria / ventas_totales

    # Promedio ponderado por categoría
    promedio_ponderado = float((promedio_por_categoria * pesos).sum())

    # Construir diccionario unificado
    calificacion_dict = promedio_por_categoria.to_dict()  # dict con {categoria: promedio}
    calificacion_dict['Promedio'] = promedio_general
    calificacion_dict['Ponderado'] = promedio_ponderado

    # Si ya existe la tienda, añadimos la clave 'calificaciones'
    if nombre_tienda in info_por_tienda:
      info_por_tienda[nombre_tienda]['calificaciones'] = calificacion_dict
    else:
      info_por_tienda[nombre_tienda] = {'calificaciones': calificacion_dict}

  else:
    # Si faltan columnas, aseguramos exista la estructura
    if nombre_tienda not in info_por_tienda:
      info_por_tienda[nombre_tienda] = {'calificaciones': {}}
    elif 'calificaciones' not in info_por_tienda[nombre_tienda]:
      info_por_tienda[nombre_tienda]['calificaciones'] = {}


# Visualizando los datos
for tienda, datos in info_por_tienda.items():
  print(f"\nCalificación promedio, promedio ponderado y promedio por categorías en {tienda}:\n")

  calif = datos.get('calificaciones', {})

  # Primero mostramos a la cabeza 'General' y 'Ponderado' siempre que existan
  for clave_fija in ['Promedio', 'Ponderado']:
    if clave_fija in calif:
      imprimir_elemento(f"-  {clave_fija}", f"-->{round(calif[clave_fija], 2): .2f}", ancho_total=40)

  # Luego el resto, excluyendo 'General' y 'Ponderado'
  resto = {k: v for k, v in calif.items() if k not in ['Promedio', 'Ponderado']}
  contador = 1
  for clave, valor in sorted(resto.items(), key=lambda x: x[1], reverse=True):
    imprimir_elemento(f"{contador}. {clave}", f"{round(valor, 2): .2f}", ancho_total=40)
    contador += 1
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
import matplotlib.pyplot as plt

for nombre_tienda, df in tiendas.items():
  if 'Producto' in df.columns:
    # Si se encuentra la columna, iniciar conteo
    producto = df['Producto'].value_counts().to_dict()

    if nombre_tienda in info_por_tienda:
      info_por_tienda[nombre_tienda]['productos'] = producto
    else:
      info_por_tienda[nombre_tienda] = {'productos': producto}
  else:
    if nombre_tienda not in info_por_tienda:
      info_por_tienda[nombre_tienda] = {}

# Crear listas para productos más y menos vendidos
productos_max = []
productos_min = []

for tienda, datos in info_por_tienda.items():
    print(f"\033[1m\nProductos más y menos populares en {tienda}:\033[0m\n")

    productos_conteo = datos.get('productos', {})
    if not 'productos':
      continue

    # Ordenar productos por cantidad (de mayor a menor)
    ordenados = sorted(productos_conteo.items(), key=lambda x: x[1], reverse=True)

    # Obtener valor máximo y mínimo
    max_cantidad = ordenados[0][1]
    min_cantidad = ordenados[-1][1]

    # Filtrar los más populares
    mas_populares = [(p, c) for p, c in ordenados if c == max_cantidad]
    menos_populares = [(p, c) for p, c in ordenados if c == min_cantidad]

    print("\tProductos más vendidos:\n")
    for i, (prod, cant) in enumerate(mas_populares, 1):
        imprimir_elemento(f"\t{i}. {prod}", cant, ancho_total=40)

    contador_p = 51
    print("\n\tProductos menos vendidos:\n")
    for i, (prod, cant) in enumerate(menos_populares, 1):
        imprimir_elemento(f"\t{contador_p}. {prod}", cant, ancho_total=40)
        contador_p -= 1

    productos_max.extend([(tienda, p, c) for p, c in mas_populares])
    productos_min.extend([(tienda, p, c) for p, c in menos_populares])

# Función para graficar
def graficar_productos(product, titulo):
  if not product:
    print(f"No hay datos para {titulo}")
    return

  etiquetas = [f"{tienda} - {prod}" for tienda, prod, _ in product]
  cantidades = [c for _, _, c in product]

  # Invertir el orden para que tienda_1 aparezca arriba
  etiquetas = etiquetas[::-1]
  cantidades = cantidades[::-1]

  plt.figure(figsize=(12, max(5, len(product)*0.5)))
  plt.barh(etiquetas, cantidades, color='#2a9d8f')
  plt.xlabel("Cantidad vendida")
  plt.title(titulo)
  plt.tight_layout()
  plt.show()


# Gráfico: productos más vendidos
graficar_productos(productos_max, "Productos más vendidos por tienda")

# Gráfico: productos menos vendidos
graficar_productos(productos_min, "Productos menos vendidos por tienda")
```
